# **Importer/Exporter Overview:**


```python
def make_top(col, top):
    plt.figure(figsize=(12, 6))
    # Get the top values
    top_values = df[col].value_counts().head(top)
    # Create a horizontal bar chart
    top_plot = top_values.plot(kind='barh', color='cyan', edgecolor='black')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.title(f'Top {top} {col} Counts')
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    plt.show()
make_top('IMPORTER COUNTRY', 10)
```

```python
top_IMPORTER_NAME = make_top('IMPORTER NAME',10)
```
